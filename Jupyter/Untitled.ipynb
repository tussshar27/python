```python
import requests
```

```python
url = "https://www.autocar.co.uk/sites/autocar.co.uk/files/styles/body-image/public/1-bmw-7-series-730ld-2019-uk-fd-hero-front.jpg?itok=3hIdcBPu"
```

```python
output = requests.get(url)
```

```python
f = open("carimage.jpg","wb")  #new file is automatically created as carimage.jpg file name, wb means write, binary format
```

```python
f.write(output.content)       #image is wrote in that file
```
